```python
# Import necessary libraries
import pandas as pd
import altair as alt
import seaborn as sns
import matplotlib.pyplot as plt
import os  # Import the os module for directory handling

# Load the dataset
file_path = 'locations.csv'  # Update this path to where your data file is located
data = pd.read_csv(file_path)

# Data Preprocessing
# Count connections per borough
borough_counts = data['boro'].value_counts().reset_index()
borough_counts.columns = ['Borough', 'Connections']

# --- Interactive Altair Bar Chart ---
borough_chart = alt.Chart(borough_counts).mark_bar().encode(
    x=alt.X('Borough', sort='-y', title='Borough'),
    y=alt.Y('Connections', title='Number of Connections'),
    color=alt.Color('Borough', legend=None, scale=alt.Scale(scheme='tableau10')),
    tooltip=['Borough', 'Connections']
).properties(
    title='Number of Connections Per Borough',
    width=600,
    height=400
)

# Save the interactive chart as an HTML file for embedding
borough_chart.save('interactive_borough_chart.html')

# --- Additional Visualization 1: Heatmap ---
# Ensure the assets directory exists
os.makedirs('assets', exist_ok=True)

plt.figure(figsize=(8, 6))
sns.heatmap(borough_counts.set_index('Borough').T, annot=True, cmap='coolwarm', fmt='g', linewidths=0.5)
plt.title('Borough Connection Heatmap')
plt.tight_layout()
# Save the heatmap image in the 'assets' folder
plt.savefig('assets/borough_heatmap.png')

# --- Additional Visualization 2: Pie Chart ---
# Pie chart of borough connections
plt.figure(figsize=(8, 6))
plt.pie(borough_counts['Connections'], labels=borough_counts['Borough'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set3', len(borough_counts)))
plt.title('Borough Connection Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
# Save the pie chart image in the 'assets' folder
plt.savefig('assets/connection_pie_chart.png')

# Display the interactive chart and contextual visualizations (this is optional for the notebook)
borough_chart.show()
```
